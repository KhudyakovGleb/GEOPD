```python
import pandas as pd
import geopandas
df = pd.read_csv(r'C:\Users\USER\Курсовая\geopandas\сохры.csv')
def geocode_address(address):
    full_address = str(address) + ", г.Санкт-Петербург, Россия"
    location = geopandas.tools.geocode(full_address, provider='nominatim', user_agent = 'my_request')
    return location['geometry'] if not location.empty else None
df['coordinates'] = df['Street'].apply(geocode_address)
import matplotlib.pyplot as plt
gdf = geopandas.GeoDataFrame(df, geometry='coordinates')
gdf.plot()
plt.show()
```

```python
output_file = 'прикол.csv'
df.to_csv(output_file, index=False)
```
